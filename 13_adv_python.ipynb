# nested if

```python
x = 3
r = x % 2

if r == 0:
    print('Even number')
    if x>5:
        print('greater number')#
        

else:
    print('Odd Number')
```

```python
x = 4
r = x % 2

if r == 0:
    print('Even number')
    if x>5:
        print('greater number')#
        

else:
    print('Odd Number')
```

```python
x = 6
r = x % 2

if r == 0:
    print('Even number')
    if x>5:
        print('greater number')
  
else:
    print('Odd Number')
```

```python
x = 2
r = x % 2

if r == 0:
    print('Even number')
    
    if x>5:
        print('greater number')
    else:
        print('not greater')
  
else:
    print('Odd Number')
```
