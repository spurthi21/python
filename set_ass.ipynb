# Set operations 

# union

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
C = {8,9,10}
```

```python
A | B
```

```python
A.union(B)
```

```python
A.union(B,C)
```

```python
A.update(B,C)
A
```

# Intersection 

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A & B
```

```python
A.intersection(B)
```

```python
A.intersection_update(B)
A
```

# difference 

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A -B  
```

```python
A.difference(B)
```

```python
B- A
```

```python
B.difference(A)
```

```python
B.difference_update(A)
B
```

# Symmetric Difference

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A^B
```

```python
A.symmetric_difference(B)
```

```python
A.symmetric_difference_update(B) 
A
```

# Subset , Superset & Disjoint

```python
A = {1,2,3,4,5,6,7,8,9}
B = {3,4,5,6,7,8}
C = {10,20,30,40}
```

```python
B.issubset(A)
```

```python
A.issuperset(B)
```

```python
C.isdisjoint(A)
```

```python
B.isdisjoint(A)
```

```python
C.issubset(A)
```

```python
A.issubset(B)
```
